```python
df_country= pd.read_excel('/content/Country-Code[1].xlsx')
df_country
```

```python
final_df=pd.merge(df,df_country, on='Country Code')
final_df.head(4)
```

```python
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts().index
country_name
```

```python
country_value=final_df.Country.value_counts().values
country_value
```

```python
plt.pie(country_value[:3],labels=country_name[:3], autopct='%1.0f%%')
```

```python
#maximum transactions of Zomato are from India
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'No.of ratings'})
ratings.head(5)
```

```python
#when ratings are betweeen 4.0 to 4.4 is very good
#when ratings are betweeen 4.5 to 4.9 is Excellent
```

```python
plt.figure(figsize=(14,10))
sns.barplot(x='Aggregate rating', y='No.of ratings', data=ratings, hue='Rating color')
```

```python
#Not rated count is very high
```

```python
plt.figure(figsize=(12,10))
sns.countplot(x='Rating color', data=ratings)
```
